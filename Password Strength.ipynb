```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('password strength data.csv',error_bad_lines=False)
df
```

```python
df.isnull().sum()
```

```python
df[df['password'].isnull()]
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(df['strength'])
```

```python
password_tuple = np.array(df)
```

```python
password_tuple
```

```python
import random
random.shuffle(password_tuple)
```

```python
x = [labels[0] for labels in password_tuple]
y = [labels[1] for labels in password_tuple]
```

```python
x
```

```python
y
```

```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
word_divide_char("kino3434")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
TfidfVectorizer()
vectorizer = TfidfVectorizer(tokenizer=word_divide_char)
X = vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_doc_vector = X[0]
first_doc_vector.T.todense()
```

```python
data = pd.DataFrame(first_doc_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
```

```python
data.sort_values(by=['TF-IDF'], ascending=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0,multi_class='multinomial')
clf.fit(X_train,y_train)
```

```python
dt = np.array(['%@123abcd'])
pred = vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
print('Accuracy = ',accuracy_score(y_test,y_pred)*100,"%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
